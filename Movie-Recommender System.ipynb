```python
##This is helper function
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i["name"])
            counter+=1
        else:
            break
    return l
```

```python
movies["cast"]=movies["cast"].apply(convert3)
```

```python
movies
```

```python
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
movies["crew"]=movies["crew"].apply(fetch_director)
```

```python
#here everything is converted in list now we are converting overview into list bcz it is a string:
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
#here we are trying to remove space between names,genre and rest of things because many times we have got the same name thats why we merge full name
```

```python
movies["genres"]=movies["genres"].apply(lambda x:[(i.replace(" ","")) for i in x])
movies["keywords"]=movies["keywords"].apply(lambda x:[(i.replace(" ","")) for i in x])
movies["cast"]=movies["cast"].apply(lambda x:[(i.replace(" ","")) for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[(i.replace(" ","")) for i in x])
```

```python
from nltk.stem.porter import PorterStemmer

# Initialize PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])  # distances with index  # second no. ke basis pr sorting kr rahe h hum 
```

```python
def recommend(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distance=similarity[movie_index ]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]  #Top 5 movies similarity
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend("Avatar")
```
